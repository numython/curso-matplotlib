## Importamos NumPy y Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Etiquetas 

### Título de la gráfica 

El título de una gráfica podemos establecerlo utilizando la función `title`

```python
x = np.linspace(0,10)
y = x * np.cos(x)
plt.plot(x, y)
plt.title("Gráfica del movimiento")
```

```python
# Modificando el estilo por defecto
x = np.linspace(0,10)
y = x * np.cos(x)
plt.plot(x, y)
plt.title("Gráfica del movimiento", fontsize=16, fontweight="bold", color="#aadd44")
```

### Etiquetas de los ejes

Las etiquetas de los ejes las colocamos utilizando las funciones `xlabel` y `ylabel` para los ejes horizontal y vertical, de manera respectiva.

```python
x = np.linspace(0,10)
y = x * np.cos(x)
plt.plot(x, y)
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (mm)")
```

```python
x = np.linspace(0,10)
y = x * np.sin(x)
plt.plot(x, y)
plt.xlabel("Tiempo (s)", color="blue", fontsize=14, 
           bbox={"boxstyle":"round", "facecolor":"#33dd33"})
plt.ylabel("Amplitud (mm)", color="red", fontsize=14, 
          bbox={"boxstyle":"round", "facecolor":"#dddd33"})
```

## Leyendas

```python
x = np.linspace(0,10)
y1 = x * np.sin(x)
y2 = x * np.cos(x)
plt.plot(x, y1, "r--", label="Partícula 1")
plt.plot(x, y2, "b--", label="Partícula 2")
plt.legend()
```

```python
x = np.linspace(0,10)
y1 = x * np.sin(x)
y2 = x * np.cos(x)
plt.plot(x, y1, "r-o")
plt.plot(x, y2, "b-x")
plt.legend(["Partícula 1", "Partícula 2"])
```

## Anotaciones

```python
x = np.linspace(0,10)
y = x * np.sin(x)
plt.plot(x, y)

plt.text(2, 4, "Anotación 1")
```

```python
x = np.linspace(0,10)
y = x * np.sin(x)
plt.plot(x, y)

plt.text(2, 4, "Anotación 1", horizontalalignment="center",
         bbox={"boxstyle":"round", "facecolor":"#ffff0077"})
```

```python
x = np.linspace(0,10)
y = x * np.sin(x)
plt.plot(x, y)
plt.annotate('Máximo local', xy=(2, 2), xytext=(2,4), 
            horizontalalignment="center", 
            arrowprops=dict(facecolor='black', shrink=0.05) )
```
